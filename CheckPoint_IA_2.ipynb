# Dataset

Importando o pandas e carregando o dataset

```python
import pandas as pd
df = pd.read_csv('/content/solicitacoescredito.csv')
df.head()
```

```python
# Vendo o tipo de dados
df.info()
```

```python
# Quantos dados únicos
df.nunique()
```

```python
# Selecionando colunas relevantes para análise
cols_analise = ['faturamentoBruto', 'endividamento', 'totalPatrimonioLiquido', 'maiorAtraso']

# Filtrando o DataFrame para incluir apenas as linhas com dados válidos nas colunas selecionadas
df_analise = df[cols_analise].dropna()

# Realizando uma análise descritiva das variáveis selecionadas
analise_descritiva = df_analise.describe()

# Calculando correlações entre as variáveis selecionadas
correlacoes = df_analise.corr()
```

```python
#Exibindo a análise desscritiva

analise_descritiva, correlacoes
```
